# 데이터프레임 구조상 행열 전환 후 분석 실시

```python
rawdata_t = rawdata.transpose()  # 행열 전환
rawdata_t
```

```python
rawdata_t.rename(columns=rawdata_t.iloc[0], inplace=True)  # 기존 columns을 첫번째 행 내용으로 일괄 변경(inplace옵션으로 해당 데이터프레임에 바로 적용)
rawdata_t.drop(rawdata_t.index[0], inplace=True)  # 첫번째 행 삭제
rawdata_t
```

```python
# 모든 종목 변화율 그래프 출력해보기
x = np.arange(len(rawdata_t.index))
top_code = []
date_arr = rawdata_t.index.tolist()

for i in range(len(rawdata_t.columns)):
    code = rawdata_t.columns[i]
    start_price = rawdata_t[code][date_arr.index("2018-06-01")]
    y = rawdata_t[code] / start_price
    plt.plot(x, y, "g-")
    if y[-1] > 8.0:
        plt.annotate(str(code),
                    xy=(x[-1], y[-1]),
                    xytext=(x[-1]+50, y[-1]),
                    arrowprops=dict(arrowstyle="->", lw=1.0, color="r"))
        top_code.append(str(code))
plt.show()
```

```python
print(top_code)  # 해당 기간(2018.6.1 ~ 2002.9.18) 상승률이 800%를 초과한 종목 Code 5개 확인
```

```python
# 상위 5개 종목 변화율 그래프 출력
x = np.arange(len(rawdata_t.index))
date_arr = rawdata_t.index.tolist()
line_color = ["pink", "green", "skyblue", "violet", "limegreen"]

for i, code in enumerate(top_code):
    start_price = rawdata_t[code][date_arr.index("2018-06-01")]
    y = rawdata_t[code] / start_price
    plt.plot(x, y, linestyle="-", color=line_color[i])
    plt.annotate(str(code),
                xy=(x[-1], y[-1]),
                xytext=(x[-1]+50, y[-1]),
                arrowprops=dict(arrowstyle="->", lw=1.0, color="r"))
plt.show()
```
