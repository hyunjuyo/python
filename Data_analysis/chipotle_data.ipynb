# 데이터 전처리

```python
chipo.info()
```

```python
chipo["item_price"].head()
```

```python
chipo["item_price"].apply(lambda x : x[1:])  # apply()함수 내에 lambda를 적용하여 첫번째 글자를 없앨 수 있음
```

##### ※ apply()함수는 시리즈 단위의 연산을 처리하는 기능을 수행하며, sum()이나 mean()과 같이 연산이 정의된 함수를 파라미터로 받음

```python
chipo["item_price"] = chipo["item_price"].apply(lambda x : float(x[1:]))  # 위 내용과 동일하게 적용하면서 해당값을 float로 변환시켜줌
```

```python
chipo.info()  # "item_price"의 타입이 변경된 것을 확인
```

```python
chipo.describe()
```

```python
chipo["order_id"] = chipo["order_id"].apply(lambda x : str(x))  # "order_id"의 타입을 str로 변환시켜줌
```

```python
chipo.info()
```

```python
chipo.describe()  # 숫자로서 계산되는 2개의 피처에 대해서만 결과값 출력되는 것을 확인함
```

# 한 주문에 30달러 이상 지불한 주문번호(id) 출력하기

```python
r = chipo.groupby("order_id")["item_price"].sum() >= 30.0  # 가격 합계 30달러 이상인 경우 확인
print(r)
```

```python
r = chipo.groupby("order_id")["item_price"].sum()
print(r)
```

```python
result = r[r >= 30]  # 위 sum()의 결과가 30 이상인 것들만 추출하기
print(result)
```

```python
result.index  # 해당 건들의 index 추출하기
```

```python
result.index.values[:20]  # 넘파이배열에 담기
```

# 각 아이템의 가격 구하기

```python
r = chipo[chipo.quantity == 1]  # "quantity"가 1인 경우만 추출
r
```

```python
result = r.groupby("item_name").min()  # 위 결과에 대해 "item_name" 그룹핑 후, 최소값 기준 추출 => 아이템별 한개씩만 출력하기 위함
result
```

```python
result.sort_values(by="item_price", ascending=False)  # "item_price"기준 내림차순으로 정렬
```

```python
x_pos = np.arange(len(result.index.tolist()))  # "item_name" 개수만큼 숫자로 넘파이배열 선언
x_pos
```

```python
price_per_item = result.item_price.tolist()  # 가격 정보를 배열에 담기
print(price_per_item)
```

```python
plt.bar(x_pos, price_per_item)
plt.ylabel("Item Price($)")
plt.title("Distribution of item price")
plt.show()
```

```python
plt.plot(x_pos, price_per_item, "go")
plt.axis([-1, 50, 0, 15])
plt.ylabel("Item Price($)")
plt.title("Distribution of item price")
plt.show()
```

```python
plt.hist(price_per_item)
plt.ylabel("Counts")
plt.title("Histogram of item price")
plt.show()
```
