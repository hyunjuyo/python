# 데이터프레임 구조상 행열 전환 후 분석 실시

```python
rawdata_tr = rawdata.transpose()  # 행열 전환
rawdata_tr
```

```python
rawdata_tr.rename(columns=rawdata_tr.iloc[0], inplace=True)  # 기존 columns을 첫번째 행 내용으로 일괄 변경(inplace옵션으로 해당 데이터프레임에 바로 적용)
rawdata_tr.drop(rawdata_tr.index[0], inplace=True)  # 첫번째 행 삭제
rawdata_tr
```

```python
# 모든 종목 변화율 그래프 출력해보기
x = np.arange(len(rawdata_tr.index))
top_code = []
date_arr = rawdata_tr.index.tolist()
start_date = "2018-06-01"

for i in range(len(rawdata_tr.columns)):
    code = rawdata_tr.columns[i]
    start_price = rawdata_tr[code][date_arr.index(start_date)]
    y = rawdata_tr[code] / start_price
    plt.plot(x, y, "g-")
    if y[-1] > 7.0:
        plt.annotate(str(code),
                    xy=(x[-1], y[-1]),
                    xytext=(x[-1]+50, y[-1]),
                    arrowprops=dict(arrowstyle="->", lw=1.0, color="r"))
        top_code.append(str(code))
plt.show()
```

```python
print(top_code)  # 해당 기간(2018.6.1 ~ 2002.9.18) 상승률 600%(7배) 초과한 종목 Code 7개 추출
```

```python
# 상위 7개 종목 변화율 그래프 출력
x = np.arange(len(rawdata_tr.index))
date_arr = rawdata_tr.index.tolist()
start_date = "2018-06-01"
line_color = ["pink", "green", "skyblue", "violet", "limegreen", "lime", "purple"]

for i, code in enumerate(top_code):
    start_price = rawdata_tr[code][date_arr.index(start_date)]
    y = rawdata_tr[code] / start_price
    plt.plot(x, y, linestyle="-", color=line_color[i])
    plt.annotate(str(code),
                xy=(x[-1], y[-1]),
                xytext=(x[-1]+50, y[-1]),
                arrowprops=dict(arrowstyle="->", lw=1.0, color="r"))
plt.show()
```

# 종목별 주가 성장률 정보 산출 및 전반적 형태 분석

```python
# 종목별 주가 성장율 정보 딕셔너리로 저장
code_list = rawdata_tr.columns.tolist()
date_list = rawdata_tr.index.tolist()
g_rate_dict = {}
start_date = "2018-06-01"
end_date = "2020-09-18"

for i in range(len(code_list)):
    code = code_list[i]
    g_rate = rawdata_tr[code][date_list.index(end_date)] / rawdata_tr[code][date_list.index(start_date)]
    g_rate_dict[code] = g_rate
```

```python
# 분석을 위해 데이터프레임 형태로 저장
g_rate_df = pd.DataFrame(data=g_rate_dict.values(), index=g_rate_dict.keys(), columns=["Growth_Rate"])
g_rate_df
```
